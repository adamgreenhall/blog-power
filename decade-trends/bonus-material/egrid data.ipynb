```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel(
    "https://www.epa.gov/sites/production/files/2020-03/egrid2018_data_v2.xlsx",
    sheet_name="GEN18",
    skiprows=1,
).set_index('SEQGEN18')
```

```python
# we see the same early-2000s natural gas boom
(df[(df.FUELG1 == 'NG') & df.GENYRRET.isnull()].groupby('GENYRONL').NAMEPCAP.sum().loc[1950:2020] / 1000).plot();
```
